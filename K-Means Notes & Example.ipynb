# K-Means 

It is a unsupervise learning algorithm, where you group imillar data into clusters.

***How it works*** :

1) choose a k value (Number of clusters).

2) The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster.

3) Until the clusters stops changing repeat the following 2 steps:
  
   Reassign data points to the cluster whose centroid is closest. Calculate new centroid of each cluster (taking mean vector).
    

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

 ## Create Data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)

# it outputs a tupple and inside that tupple there is numpy arrays
# the data has 200 samples and 2 features being clusterd into 4 clusters or blobbs

#data[0].shape 

# The 2nd item in the tupple, it conatins the cluster that it belongs to [0,1,2,3]
data[1]
```

## Visulize Data

```python
# Plot x as the 1st column of the 1st item & y as the 2nd column of the 1st item
# c is the color so we have 4 clusters it will make each diff color

plt.scatter(data[0][:,0] , data[0][:,1], c = data[1])
```

## K-Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
# return the center of each cluster in a numpy array
kmeans.cluster_centers_
```

```python
data[1]
```

```python
# predicted labels 
kmeans.labels_
```

```python
# if we have un labeled data we wont comapre the labels but here we have the predicted and the real labels
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6)) # sharey that they share axis
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
# plotting the centroids of each cluster

plt.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.show()
```

## Find The Best K Value Using Elbow Method

```python
# Use elbow method to find the ideal number of clusters
best = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(data[0])
    best.append(np.sqrt(kmeans.inertia_))

plt.plot(range(1, 11), best, marker='s');
plt.xlabel('$k$')
plt.ylabel('$best$');
```
